```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
data = pd.read_csv('Electricity_P.csv', engine='python', usecols=['UNIX_TS', 'WHE', 'MHE'], encoding='CP949')
df=pd.read_csv('Climate_HourlyWeather.csv', engine='python', usecols=['Date/Time', 'Temp (C)', 'Rel Hum (%)', 'Weather'], encoding='CP949')
data.head()
```

```python
df.head()
```

```python
#Temp(C), Rel_Hum(%)
data.rename(columns={'UNIX_TS': 'Time'}, inplace=True)
df.rename(columns={'Date/Time': 'key', 'Temp (C)': 'Temp', 'Rel Hum (%)': 'Rel_Hum'}, inplace=True)
```

```python
data['Time'] = pd.to_datetime(data['Time'], unit='s')
df['key'] = pd.to_datetime(df['key'], infer_datetime_format=True)
data.head()
```

```python
df.head()
```

```python
df['Weather'].mask(df['Weather'] == "Heavy Rain - Fog", 0, inplace=True)
df['Weather'].mask(df['Weather'] == "Heavy Rain", 1, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain Showers - Fog", 2, inplace=True)
df['Weather'].mask(df['Weather'] == "Thunderstorms - Rain Showers", 2, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain Showers", 3, inplace=True)
df['Weather'].mask(df['Weather'] == "Snow Showers", 3, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain Showers - Snow Showers", 3, inplace=True)
df['Weather'].mask(df['Weather'] == "Moderate Rain Showers", 4, inplace=True)
df['Weather'].mask(df['Weather'] == "Freezing Rain - Snow Showers", 3, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain - Fog - Blowing Dust", 5, inplace=True)
df['Weather'].mask(df['Weather'] == "Thunderstorms - Rain", 5, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain - Fog", 5, inplace=True)
df['Weather'].mask(df['Weather'] == "Snow - Fog", 5, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain", 6, inplace=True)
df['Weather'].mask(df['Weather'] == "Snow", 6, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain - Snow", 6, inplace=True)
df['Weather'].mask(df['Weather'] == "Moderate Snow", 7, inplace=True)
df['Weather'].mask(df['Weather'] == "Moderate Rain - Fog", 6, inplace=True)
df['Weather'].mask(df['Weather'] == "Thunderstorms - Moderate Rain", 6, inplace=True)
df['Weather'].mask(df['Weather'] == "Freezing Rain", 6, inplace=True)
df['Weather'].mask(df['Weather'] == "Moderate Rain", 7, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain - Drizzle - Fog", 6, inplace=True)
df['Weather'].mask(df['Weather'] == "Rain - Drizzle", 7, inplace=True)
df['Weather'].mask(df['Weather'] == "Snow - Snow Grains", 7, inplace=True)
df['Weather'].mask(df['Weather'] == "Drizzle - Fog", 7, inplace=True)
df['Weather'].mask(df['Weather'] == "Freezing Drizzle - Fog", 7, inplace=True)
df['Weather'].mask(df['Weather'] == "Drizzle - Snow Grains - Fog", 7, inplace=True)
df['Weather'].mask(df['Weather'] == "Drizzle", 8, inplace=True)
df['Weather'].mask(df['Weather'] == "Snow Grains", 8, inplace=True)
df['Weather'].mask(df['Weather'] == "Moderate Drizzle - Fog", 8, inplace=True)
df['Weather'].mask(df['Weather'] == "Freezing Drizzle", 8, inplace=True)
df['Weather'].mask(df['Weather'] == "Fog", 9, inplace=True)
df['Weather'].mask(df['Weather'] == "Thunderstorms", 9, inplace=True)
df['Weather'].mask(df['Weather'] == "Freezing Fog", 9, inplace=True)
df['Weather'].mask(df['Weather'] == "Cloudy", 10, inplace=True)
df['Weather'].mask(df['Weather'] == "Mostly Cloudy", 11, inplace=True)
df['Weather'].mask(df['Weather'] == "Mainly Clear", 12, inplace=True)
df['Weather'].mask(df['Weather'] == "Clear", 13, inplace=True)
df
```

```python
data['key'] = data.Time.dt.strftime('%d/%m/%y %H')
df['key'] = df.key.dt.strftime('%d/%m/%y %H')
data.head()
```

```python
df.head()
```

```python
res_df = pd.merge(data, df, on="key")
res_df
```

```python
res_df = res_df.replace(' ', '')
res_df = res_df.replace('', np.nan)
res_df = res_df.dropna()
res_df.isnull().sum()
```

```python
res_df['Hour'] = res_df.Time.dt.strftime('%H')
res_df['Month'] = res_df.Time.dt.strftime('%m')
res_df
```

```python
res_df.info()
```

```python
res_df.to_csv("res_data_min.csv", mode='w', columns=['Time', 'WHE', 'MHE', 'Temp', 'Rel_Hum', 'Weather', 'Hour', 'Month'], index=False)
print('end')
```
