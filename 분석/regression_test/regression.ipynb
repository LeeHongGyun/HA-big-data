```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
data = pd.read_csv('data/res_data_hour.csv', engine='python', encoding='CP949')
data.head()
```

```python
model = smf.ols(formula = 'WHE ~ Temp + Rel_Hum + Weather + Hour + Month', data = data)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = 'WHE ~ Temp + Weather + Hour + Month', data = data)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = 'WHE ~ Rel_Hum + Weather + Hour', data = data)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = 'WHE ~ Temp + Rel_Hum + Hour', data = data)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = 'WHE ~ Temp + Hour + Month', data = data)
result = model.fit()
result.summary()
```
