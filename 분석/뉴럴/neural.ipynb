```python
import tensorflow as tf
import numpy as np
import pandas as pd

tf.set_random_seed(777)
data = pd.read_csv('data/res_data_min.csv', engine='python', encoding='CP949')

x_data = data[['Temp', 'Rel_Hum', 'Weather', 'Month']]
y_data = data['WHE']
print("end")
x_data.dtypes
```

```python
X = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)

w = tf.Variable(tf.random_normal([4, 1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")

hypothesis = tf.matmul(X, w) + b

cost = tf.reduce_mean(tf.square(hypothesis - y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate =1e-5)

train = optimizer.minimize(cost)

print("end")
```

```python
sess = tf.Session()

sess.run(tf.global_variables_initializer())

for step in range (2001):
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train],
        feed_dict = {X:x_data, y:y_data})
    if step % 10 == 0:
        print(step, "cost: ", cost_val, "\n prediction: ", hy_val)
```

```python
print("temp, hum, weather, month")
print('4.7, 79.0, 11.0, 04, 918: ', sess.run(hypothesis, feed_dict={X:[[4.7, 79.0, 11.0, 4]]}))
print('5.6, 77.0, 11.0, 02, 1096', sess.run(hypothesis, feed_dict={X:[[5.6, 77.0, 11.0, 2]]}))
print('10.3, 76.0, 12.0, 10, 1596', sess.run(hypothesis, feed_dict={X:[[10.3, 76.0, 12.0, 10]]}))
print('14.8, 69.0, 10.0, 06, 781', sess.run(hypothesis, feed_dict={X:[[14.8, 69.0, 10.0, 6]]}))
```
